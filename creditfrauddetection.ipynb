```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("C:\\Users\\Dell\\Downloads\\archive (2)\\creditcard.csv")
```

```python
df
```

```python
#Data preprocessing
#Handle missing values
```

```python
#check for missing values
print(df.isnull().sum())
```

```python
df=df.dropna()
```

```python
# Features and labels
X = df.drop(columns=['Class', 'Time', 'Amount'])  # Dropping 'Time' and 'Amount' for normalization separately
y = df['Class']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Normalize 'Amount' feature
scaler = StandardScaler()
df['Amount'] = scaler.fit_transform(df[['Amount']])
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Train Logistic Regression model with class_weight='balanced'
lr_model = LogisticRegression(class_weight='balanced', random_state=42)
lr_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Predict on the test set
y_pred_lr = lr_model.predict(X_test)
```

```python
# Evaluate model performance
print("Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred_lr))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Train Random Forest model with class_weight='balanced'
rf_model = RandomForestClassifier(class_weight='balanced', random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred_rf = rf_model.predict(X_test)
```

```python
# Evaluate Random Forest model
print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))
```

```python
# Confusion matrix
print("Confusion Matrix (Logistic Regression):")
print(confusion_matrix(y_test, y_pred_lr))
```

```python
print("Confusion Matrix (Random Forest):")
print(confusion_matrix(y_test, y_pred_rf))
```
